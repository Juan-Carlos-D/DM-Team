### Outlier Detection and Removal

To further refine the dataset and ensure that **extreme values do not distort the clustering**, we conducted an **outlier detection process** using the **Interquartile Range (IQR) method**. Outliers were identified for each selected numerical feature based on values that fell:  
- **Below** Q1 - 1.5 × IQR  
- **Above** Q3 + 1.5 × IQR  

Where **Q1 and Q3** represent the first and third quartiles, respectively.


### Outliers Identified

The analysis examined the following variables for potential outliers:
- `cases_per_100k`
- `deaths_per_100k`
- `poverty`
- `median_income`

Outliers in these variables may reflect:
- **Extreme socioeconomic disparities**
- **Unusually high or low case/death rates**

### Outlier Removal

To maintain the **integrity of the dataset**, we removed rows containing outliers in any of the variables listed above. As a result:
- Rows with extreme values were excluded from further analysis.
- The dataset now reflects a **more balanced and representative sample** of Texas counties for clustering.

This step improves **data quality** and ensures that clustering is **not biased by a few extreme outliers**.

````r
# Define function to count outliers using IQR
count_outliers <- function(x) {
  Q1 <- quantile(x, 0.25, na.rm = TRUE)
  Q3 <- quantile(x, 0.75, na.rm = TRUE)
  IQR_value <- Q3 - Q1
  sum(x < (Q1 - 1.5 * IQR_value) | x > (Q3 + 1.5 * IQR_value), na.rm = TRUE)
}

# Apply to all numerical variables used in clustering (excluding 'county')
outlier_summary <- covid_census_tx %>%
  select(-county) %>%
  summarize(across(everything(), count_outliers, .names = "{.col}_outliers")) %>%
  pivot_longer(cols = everything(), names_to = "Variable", values_to = "Outlier_Count")

# Display outlier summary
print(outlier_summary, width = Inf)
````

````r
# Custom function to check if value is NOT an outlier using IQR
is_not_outlier <- function(x) {
  Q1 <- quantile(x, 0.25, na.rm = TRUE)
  Q3 <- quantile(x, 0.75, na.rm = TRUE)
  IQR_value <- Q3 - Q1
  x >= (Q1 - 1.5 * IQR_value) & x <= (Q3 + 1.5 * IQR_value)
}

# Remove rows containing outliers from key analysis variables
covid_census_tx_cleaned <- covid_census_tx %>%
  filter(
    is_not_outlier(cases_per_100k) &
    is_not_outlier(deaths_per_100k) &
    is_not_outlier(total_pop) &
    is_not_outlier(median_income) &
    is_not_outlier(commuters_by_public_transportation) &
    is_not_outlier(pct_on_food_stamps) &
    is_not_outlier(pct_work_from_home) &
    is_not_outlier(poverty) &
    is_not_outlier(income_per_capita) &
    is_not_outlier(gini_index)
  )

# Calculate and display how many rows were removed
rows_removed <- nrow(covid_census_tx) - nrow(covid_census_tx_cleaned)
cat("Number of rows removed due to outliers:", rows_removed, "\n")
````

### 3.7 K-Means Clustering Method

To identify patterns across Texas counties, we applied the **K-means clustering algorithm**, a popular unsupervised machine learning method used to partition data into distinct groups.

#### How K-means Works:
- Assigns each data point to the **nearest centroid**, forming K clusters
- Iteratively updates centroids to minimize the **within-cluster sum of squares (WSS)**
- Sensitive to scale, so we standardized all input variables prior to clustering

#### Why We Use It:
K-means is well-suited for this project because it:
- Finds natural groupings in multivariate data
- Is efficient on moderate-sized datasets
- Provides clear cluster centers for profile interpretation

We tested both **K = 3** and **K = 4** clusters to explore how different levels of granularity affect cluster quality and interpretability. The results are visualized below.

````r
# K-means with k = 3
set.seed(42)
kmeans_3 <- kmeans(scaled_df, centers = 3, nstart = 25)

# K-means with k = 4
set.seed(42)
kmeans_4 <- kmeans(scaled_df, centers = 4, nstart = 25)
````

### 3.9 Hierarchical Clustering Method

In addition to K-means, we applied **hierarchical clustering** to better understand the nested structure of counties and explore whether similar patterns emerged using a different unsupervised learning approach.

#### How Hierarchical Clustering Works:
- Builds a tree (dendrogram) by iteratively merging (agglomerative) or splitting (divisive) clusters
- Does not require pre-specifying the number of clusters
- Can use various linkage methods to determine cluster distance

#### Methods Used:
We tested two agglomerative linkage strategies:
- **Ward’s Method**: Minimizes total within-cluster variance (produces more compact, spherical clusters)
- **Complete Linkage**: Merges clusters based on the *maximum* distance between observations (tends to produce elongated, well-separated clusters)

## 3.16 Unsupervised Evaluation of Clustering Quality

In this section, we evaluate the quality of our clustering solutions **without using any ground truth labels**, focusing on internal validation.

### Method Used: Elbow Method (WSS)

The **Elbow Method** uses the **Within-Cluster Sum of Squares (WSS)** to assess cluster compactness. For each value of `k` (number of clusters), WSS measures how tightly grouped the data points are within each cluster.

- A **lower WSS** indicates more compact clusters.
- As `k` increases, WSS naturally decreases.
- The **“elbow” point** on the plot represents the value of `k` where adding more clusters yields **diminishing returns** in terms of WSS reduction.

This helps us **choose an optimal number of clusters** by balancing compactness and simplicity. In our analysis, we identified `k = 3` as a reasonable choice based on the elbow shape in the WSS plot.

This unsupervised evaluation step strengthens the justification for selecting three clusters in both K-means and hierarchical clustering approaches.

### Unsupervised Evaluation: Silhouette Analysis

To further assess cluster validity, we use **silhouette scores**, which measure how similar each point is to its own cluster versus other clusters. A higher silhouette score indicates that the observation is well matched to its own cluster and poorly matched to neighboring clusters.

#### **K-means (k = 3)**
- **Average silhouette width**:  
  - Cluster 1: **0.21**  
  - Cluster 2: **0.17**  
  - Cluster 3: **0.16**

- K-means shows relatively low but positive silhouette scores, with all clusters above 0.15, suggesting **some separation**, though not perfect.

#### **Hierarchical Clustering (Ward, k = 3)**
- **Average silhouette width**:  
  - Cluster 1: **0.13**  
  - Cluster 2: **0.19**  
  - Cluster 3: **0.40**

- One Ward cluster (Cluster 3) shows a **much stronger silhouette (0.40)**, indicating tight internal cohesion and good separation. However, Cluster 1 performs **worse** than in K-means.

### Interpretation

- **K-means** provides more **balanced silhouette scores** across all three clusters, reinforcing its use as a reliable baseline.

## What is DBSCAN?

**DBSCAN** (Density-Based Spatial Clustering of Applications with Noise) is a powerful clustering algorithm that groups together data points that are **closely packed** and labels points in low-density regions as **outliers (noise)**.

Unlike K-means or hierarchical clustering, DBSCAN:
- **Does not require specifying the number of clusters in advance**
- **Can find arbitrarily shaped clusters**
- **Handles noise and outliers naturally**

### Key Concepts

- **Core Point**: A point with at least `minPts` neighbors within a radius `eps`.
- **Border Point**: A point that is within `eps` of a core point but doesn't have enough neighbors to be a core point itself.
- **Noise Point**: A point that is not a core point and not within `eps` of any core point.

### Key Parameters

- `eps`: The maximum distance between two points to be considered neighbors.
- `minPts`: The minimum number of points required to form a dense region (including the point itself).

A good rule of thumb:
- Use **domain knowledge or a k-distance plot** to determine `eps`
- Set `minPts = dimensionality + 1` (e.g., if 10 variables, try minPts = 11)

### Why Use DBSCAN?

- Excellent for **real-world datasets** where clusters are not well-separated or spherical.
- Automatically detects **noise** (e.g., counties that don’t clearly fit a pattern).
- Ideal for **spatial data**, such as geographic county-level COVID patterns, where population clusters may vary in shape and density.

## Gaussian Mixture Models (GMM) — Clustering Explanation

A **Gaussian Mixture Model (GMM)** is a *probabilistic clustering technique* that assumes the data is generated from a mixture of several **Gaussian (normal) distributions** with unknown parameters.

Unlike **K-means**, which assigns each point to a *single cluster* (hard assignment), GMM assigns **probabilities** to each point for belonging to each cluster (soft assignment). This allows for greater flexibility in modeling **elliptical or overlapping clusters**.

#### Key Features

- **Probabilistic Assignment**: Each data point has a probability of belonging to every cluster.
- **Soft Clustering**: Accounts for uncertainty in cluster membership.
- **Flexible Cluster Shapes**: Can model clusters of different shapes, sizes, and densities.
- **Expectation-Maximization (EM)**: Algorithm used to estimate the parameters (mean, covariance, and mixture weight) of each Gaussian component.

#### When to Use GMM

- When clusters are *not clearly separated or spherical*.
- When **K-means assumptions are too rigid**.
- When *soft assignments* are preferable to hard ones.
- When data likely contains **overlapping distributions**.

### Understanding BIC in Gaussian Mixture Models (GMM)

The **Bayesian Information Criterion (BIC)** is a model selection metric used to evaluate the trade-off between model fit and complexity. In the context of **Gaussian Mixture Models**, BIC helps determine:

- The **optimal number of clusters (components)**.
- The most appropriate **covariance structure** (e.g., spherical, diagonal, ellipsoidal).

#### How BIC Works:
- **Lower BIC values** indicate better models.
- BIC penalizes model complexity, so even if a model fits the data better, it must do so **efficiently** to be preferred.
- GMM models are compared across a range of component numbers and structures, and the best model minimizes BIC.
