### 2.3.3 Outlier Detection and Removal

To further refine the dataset and ensure that **extreme values do not distort the clustering**, we conducted an **outlier detection process** using the **Interquartile Range (IQR) method**. Outliers were identified for each selected numerical feature based on values that fell:  
- **Below** Q1 - 1.5 × IQR  
- **Above** Q3 + 1.5 × IQR  

Where **Q1 and Q3** represent the first and third quartiles, respectively.


### Outliers Identified

The analysis examined the following variables for potential outliers:
- `cases_per_100k`
- `deaths_per_100k`
- `poverty`
- `median_income`
- `pct_on_food_stamps`
- `commuters_by_public_transportation`
- `pct_work_from_home`
- `income_per_capita`
- `gini_index`
- `total_pop`

Outliers in these variables may reflect:
- **Extreme socioeconomic disparities**
- **Unusually high or low case/death rates**
- **Reporting inconsistencies**

### Outlier Removal

To maintain the **integrity of the dataset**, we removed rows containing outliers in any of the variables listed above. As a result:
- Rows with extreme values were excluded from further analysis.
- The dataset now reflects a **more balanced and representative sample** of Texas counties for clustering.

This step improves **data quality** and ensures that clustering is **not biased by a few extreme outliers**.

````r
# Define function to count outliers using IQR
count_outliers <- function(x) {
  Q1 <- quantile(x, 0.25, na.rm = TRUE)
  Q3 <- quantile(x, 0.75, na.rm = TRUE)
  IQR_value <- Q3 - Q1
  sum(x < (Q1 - 1.5 * IQR_value) | x > (Q3 + 1.5 * IQR_value), na.rm = TRUE)
}

# Apply to all numerical variables used in clustering (excluding 'county')
outlier_summary <- covid_census_tx %>%
  select(-county) %>%
  summarize(across(everything(), count_outliers, .names = "{.col}_outliers")) %>%
  pivot_longer(cols = everything(), names_to = "Variable", values_to = "Outlier_Count")

# Display outlier summary
print(outlier_summary, width = Inf)
````

````r
# Custom function to check if value is NOT an outlier using IQR
is_not_outlier <- function(x) {
  Q1 <- quantile(x, 0.25, na.rm = TRUE)
  Q3 <- quantile(x, 0.75, na.rm = TRUE)
  IQR_value <- Q3 - Q1
  x >= (Q1 - 1.5 * IQR_value) & x <= (Q3 + 1.5 * IQR_value)
}

# Remove rows containing outliers from key analysis variables
covid_census_tx_cleaned <- covid_census_tx %>%
  filter(
    is_not_outlier(cases_per_100k) &
    is_not_outlier(deaths_per_100k) &
    is_not_outlier(total_pop) &
    is_not_outlier(median_income) &
    is_not_outlier(commuters_by_public_transportation) &
    is_not_outlier(pct_on_food_stamps) &
    is_not_outlier(pct_work_from_home) &
    is_not_outlier(poverty) &
    is_not_outlier(income_per_capita) &
    is_not_outlier(gini_index)
  )

# Calculate and display how many rows were removed
rows_removed <- nrow(covid_census_tx) - nrow(covid_census_tx_cleaned)
cat("Number of rows removed due to outliers:", rows_removed, "\n")
````

Additionally, understanding the spread and skewness in variables such as `deaths_per_100k`, `poverty`, and `median_income` will guide preprocessing decisions like normalization or transformation prior to model fitting.

````r
# Function to compute key statistics for census data
compute_census_stats <- function(data, variables) {
  data %>%
    summarize(across(
      all_of(variables),
      list(
        Range = ~ max(.x, na.rm = TRUE) - min(.x, na.rm = TRUE),
        Mode = ~ mfv(.x)[1],
        Mean = ~ mean(.x, na.rm = TRUE),
        Median = ~ median(.x, na.rm = TRUE),
        Variance = ~ var(.x, na.rm = TRUE),
        SD = ~ sd(.x, na.rm = TRUE)
      ),
      .names = "{.col}_{.fn}"
    ))
}

# Define key census variables for analysis
census_variables <- c("cases_per_100k", "deaths_per_100k", "poverty", "median_income", 
                      "pct_on_food_stamps", "commuters_by_public_transportation",
                      "pct_work_from_home", "income_per_capita", 
                      "gini_index", "total_pop")

# Compute statistics
census_stats_table <- compute_census_stats(covid_census_tx_cleaned, census_variables)

# Convert to long format for better readability
census_stats_long <- census_stats_table %>%
  pivot_longer(
    cols = everything(),
    names_to = "Combined",
    values_to = "Value"
  ) %>%
  mutate(
    Variable = str_extract(Combined, ".*(?=_)"),  # Extract everything before the last underscore
    Statistic = str_extract(Combined, "[^_]+$")  # Extract everything after the last underscore
  ) %>%
  select(Variable, Statistic, Value) %>%
  pivot_wider(
    names_from = "Statistic",
    values_from = "Value",
    values_fn = function(x) x[1]  # Ensure no list-columns
  ) %>%
  # Round all numeric columns to 2 decimals
  mutate(across(where(is.numeric), ~ round(.x, 2))) %>%
  # Format large numbers with commas
  mutate(across(where(is.numeric), ~ format(.x, big.mark = ",", scientific = FALSE)))  

# Disable scientific notation globally
options(scipen = 999)

# Display as interactive table
datatable(
  census_stats_long,
  rownames = FALSE,
  options = list(
    pageLength = 10,
    autoWidth = TRUE
  ),
  caption = "Descriptive Statistics for Selected Variables"
)
````

### 3.3 K-Means Clustering Method

To identify patterns across Texas counties, we applied the **K-means clustering algorithm**, a widely used unsupervised machine learning technique for discovering structure in multivariate data.

#### How K-means Works:
- Assigns each data point to the **nearest centroid**, grouping observations into K clusters
- Iteratively updates the cluster centroids to minimize the **within-cluster sum of squares (WSS)**
- Is sensitive to the scale of variables, which is why we standardized all input features prior to clustering

#### Why We Use It:
K-means is well-suited for this project because it:
- Efficiently detects **natural groupings** in complex datasets
- Performs well on moderate-sized data like our county-level inputs
- Produces **interpretable centroids**, enabling us to profile each cluster based on standardized feature values

To evaluate how the number of clusters affects interpretability and structure, we tested both **K = 3** and **K = 4** cluster solutions. These results are visualized in the next section to help compare structure and separation across different values of K.

### 3.5 Hierarchical Clustering and Dendrogram Interpretation

In addition to K-means, we applied **hierarchical clustering** to explore nested relationships between Texas counties and evaluate the robustness of our findings using an alternative unsupervised learning approach.

#### How Hierarchical Clustering Works:
- Builds a **dendrogram (tree structure)** by iteratively merging smaller clusters into larger ones (agglomerative) or splitting larger ones into smaller ones (divisive)
- Does **not require pre-specifying** the number of clusters
- Allows flexibility through different **linkage strategies** that define how cluster distances are calculated

#### Linkage Methods Used:
We evaluated two agglomerative methods:

- **Ward’s Method**: Minimizes within-cluster variance, producing **compact and spherical clusters**. It often performs similarly to K-means and is easy to interpret.
- **Complete Linkage**: Merges clusters using the **maximum distance** between observations, resulting in **elongated and more spread-out clusters**.

The dendrograms below show how counties were grouped using both linkage types, with rectangles highlighting the **K = 3** solution for comparison with K-means.

#### Interpretation:

- **Ward’s Method**:
  - Produces **balanced, cohesive clusters** with even branch lengths
  - Strong alignment with the **K-means structure**
  - Useful for identifying compact, well-defined groupings

- **Complete Linkage**:
  - Results in **more variable cluster sizes** and **less compact branches**
  - Captures **outliers and edge cases** more clearly
  - Visual separation is greater, but interpretability is more challenging

### Conclusion

Both methods suggest that **three clusters** is a natural grouping in the data. While Complete Linkage highlights more variability, **Ward’s Method is more consistent with the K-means results** and easier to interpret.

Hierarchical clustering provides **additional validation** of our groupings and suggests that certain counties cluster together consistently regardless of algorithm — reinforcing the **stability and reliability** of our findings.

````r
# Hierarchical clustering - Ward
hc_ward <- hclust(dist(scaled_df), method = "ward.D2")

# Hierarchical clustering - Complete Linkage
hc_complete <- hclust(dist(scaled_df), method = "complete")
````

### 3.7 Methodology: Ward Hierarchical Clustering

To segment Texas counties based on their socioeconomic and COVID-related characteristics, we applied **Ward’s hierarchical clustering algorithm** using the following process:

- **Feature Selection**: We selected 10 key numeric variables that reflect county-level socioeconomic status, public health risk, and pandemic impact. These included `poverty`, `median_income`, `income_per_capita`, `public transportation use`, `remote work capability`, and `COVID-19 cases and deaths per 100k`.

- **Standardization**: All selected variables were standardized using z-score scaling to ensure that each feature contributes equally to the clustering process, regardless of original units or range.

- **Distance Matrix**: We computed a Euclidean distance matrix between counties based on the standardized features, capturing the dissimilarity between counties.

- **Clustering Algorithm**: We applied **Ward’s method** (`ward.D2`), a hierarchical agglomerative clustering technique that merges clusters to minimize the total within-cluster variance. This method is effective at forming compact, spherical clusters.

- **Cluster Assignment**: We cut the resulting dendrogram to create **three clusters (k = 3)**. These groupings were then added back to the dataset for profiling and visualization.

This methodology ensures that counties are grouped based on holistic, multi-dimensional similarity—balancing economic, demographic, and pandemic-related factors.

````r
# Numeric columns for distance matrix
numeric_columns <- c("cases_per_100k", "deaths_per_100k", "poverty", "median_income",
                     "pct_on_food_stamps", "commuters_by_public_transportation",
                     "pct_work_from_home", "income_per_capita", "gini_index", "total_pop")

# Create distance matrix
dist_matrix <- dist(scale(covid_census_tx_cleaned[, numeric_columns]))

# Perform ward clustering
ward_clusters <- cutree(hclust(dist_matrix, method = "ward.D2"), k = 3)

# Add Ward cluster labels to dataset
covid_census_tx_ward <- covid_census_tx_cleaned %>%
  mutate(ward_cluster = factor(ward_clusters))
````

## What is DBSCAN?

**DBSCAN** (Density-Based Spatial Clustering of Applications with Noise) is a powerful clustering algorithm that groups together data points that are **closely packed** and labels points in low-density regions as **outliers (noise)**.

Unlike K-means or hierarchical clustering, DBSCAN:
- **Does not require specifying the number of clusters in advance**
- **Can find arbitrarily shaped clusters**
- **Handles noise and outliers naturally**

### Key Concepts

- **Core Point**: A point with at least `minPts` neighbors within a radius `eps`.
- **Border Point**: A point that is within `eps` of a core point but doesn't have enough neighbors to be a core point itself.
- **Noise Point**: A point that is not a core point and not within `eps` of any core point.

### Key Parameters

- `eps`: The maximum distance between two points to be considered neighbors.
- `minPts`: The minimum number of points required to form a dense region (including the point itself).

A good rule of thumb:
- Use **domain knowledge or a k-distance plot** to determine `eps`
- Set `minPts = dimensionality + 1` (e.g., if 10 variables, try minPts = 11)

### Why Use DBSCAN?

- Excellent for **real-world datasets** where clusters are not well-separated or spherical.
- Automatically detects **noise** (e.g., counties that don’t clearly fit a pattern).
- Ideal for **spatial data**, such as geographic county-level COVID patterns, where population clusters may vary in shape and density.

## 3.11 Gaussian Mixture Models (GMM)

### Overview

A **Gaussian Mixture Model (GMM)** is a *probabilistic clustering technique* that assumes the data is generated from a mixture of several **Gaussian (normal) distributions**, each with unknown parameters.

Unlike **K-means**, which assigns each point to a single cluster (**hard assignment**), GMM assigns a **probability** to each point for belonging to each cluster (**soft assignment**). This provides greater flexibility in modeling **elliptical**, **uneven**, or **overlapping clusters**.

#### Key Features of GMM

- **Probabilistic Assignment**: Each data point receives a probability score for each cluster.
- **Soft Clustering**: Handles uncertainty in cluster membership.
- **Flexible Cluster Shapes**: Can model clusters of various shapes and sizes.
- **Expectation-Maximization (EM)**: An iterative algorithm used to estimate cluster parameters (mean, covariance, and mixing proportions).

#### When to Use GMM

- When clusters are not clearly separated or are **elliptical** rather than spherical.
- When **overlap between groups** is likely.
- When **soft assignments** are more informative than hard labels.
- When **K-means assumptions** (equal-sized, spherical clusters) do not hold.

### 3.12.3 Understanding BIC in Gaussian Mixture Models (GMM)

The **Bayesian Information Criterion (BIC)** is a model selection metric used to evaluate the trade-off between model fit and complexity. In the context of **Gaussian Mixture Models**, BIC helps determine:

- The **optimal number of clusters (components)**.
- The most appropriate **covariance structure** (e.g., spherical, diagonal, ellipsoidal).

#### How BIC Works:
- **Lower BIC values** indicate better models.
- BIC penalizes model complexity, so even if a model fits the data better, it must do so **efficiently** to be preferred.
- GMM models are compared across a range of component numbers and structures, and the best model minimizes BIC.
